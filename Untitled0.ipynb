```python
import pandas as pd
import sqlite3 as sq
```

```python
loan = pd.read_csv('/content/drive/MyDrive/work/loan.csv', low_memory = False)
loan = loan[loan['loan_status']!= 'Current']
print(loan['loan_status'].value_counts())
print(loan.shape)
```

```python
at_risk = ['Charged Off', 'Late (31-120 days)', 'Late (16-30 days)', 'In Grace Period', 'Does not meet the credit policy']
loan['target'] = loan['loan_status'].isin(at_risk).astype(int)
print(loan['target'].value_counts())
```

```python
leakage_cols = ['recoveries', 'collection_recovery_fee', 'total_rec_late_fee', 'total_pymnt', 'total_pymnt_inv', 'total_rec_prncp', 'total_rec_int', 'out_prncp', 'out_prncp_inv', 'last_pymnt_amnt', 'last_pymnt_d', 'next_pymnt_d', 'debt_settlement_flag', 'settlement_status', 'settlement_amount', 'settlement_percentage', 'settlement_term', 'debt_settlement_flag_date']
irrelevant_cols = ['id','member_id','url','desc','title','zip_code','emp_title']
hardship_cols = [col for col in loan.columns if col.startswith('hardship_')]

all_potential_cols_todrop = leakage_cols + irrelevant_cols + hardship_cols

# Filter out columns that do not exist in the DataFrame
cols_todrop = [col for col in all_potential_cols_todrop if col in loan.columns]

loan = loan.drop(columns=cols_todrop)
print(loan.shape)
```

```python
missing = loan.isnull().sum()
missing_pct = (missing/len(loan))*100
print(missing_pct[missing_pct>0].sort_values(ascending = False))
```

```python
high_missing = missing_pct[missing_pct>40].index.tolist()
loan = loan.drop(columns = high_missing)
print(loan.shape)
```

```python
# Identify numerical and categorical column names
numerical_col_names = loan.select_dtypes(include=['number']).columns
categorical_col_names = loan.select_dtypes(include=['object', 'category']).columns

# Fill numerical columns using their median
# Using .loc to ensure modification of the original DataFrame
loan.loc[:, numerical_col_names] = loan[numerical_col_names].fillna(loan[numerical_col_names].median())

# Fill categorical columns using their mode
# Using .loc to ensure modification of the original DataFrame
# .mode() returns a DataFrame, so we use .iloc[0] to get the first row of modes
loan.loc[:, categorical_col_names] = loan[categorical_col_names].fillna(loan[categorical_col_names].mode().iloc[0])

print(loan.isnull().sum().sum())
```

```python
conn = sq.connect('loans.db')
loan.to_sql('loans',conn,if_exists ='replace', index = False)
result = pd.read_sql_query("SELECT COUNT(*) FROM loans",conn)
print(result)
```

```python
query1 = """
SELECT grade,
       COUNT(*) as total_loans,
       SUM(target) as defaults,
       ROUND(AVG(target)*100,2) as default_rate_pct
FROM loans
GROUP BY grade
ORDER BY grade
"""
print(pd.read_sql_query(query1,conn))
```

```python
query2 = """
SELECT purpose,
       COUNT(*) as total_loans,
       SUM(target) as defaults,
       ROUND(AVG(target)*100,2) as default_rate_pct
FROM loans
GROUP BY purpose
ORDER BY purpose
"""
print(pd.read_sql_query(query2,conn))
```

```python
query3= """
SELECT target,
      AVG(int_rate) as avg_int_rate,
      AVG(annual_inc) as avg_annual_inc,
      AVG(dti) as avg_dti
FROM loans
GROUP BY target
"""
print(pd.read_sql_query(query3,conn))
```

```python
query4= """
SELECT
      CASE
          WHEN dti<10 THEN 'low(0-10)'
          WHEN dti BETWEEN 10 AND 20 THEN 'Medium (10-20)'
          WHEN dti BETWEEN 20 AND 30 THEN 'High(20-30)'
          ELSE 'Very High(30+)'
      END as dti_bucket,
      COUNT(*) as total_loans,
      ROUND(AVG(target)*100,2) as default_rate_pct
FROM loans
GROUP BY dti_bucket
ORDER BY default_rate_pct
"""
print(pd.read_sql_query(query4,conn))
```

```python
cols_todrop_again = ['loan_status','issue_d','earliest_cr_line','last_credit_pull_d']
loan = loan.drop(columns = cols_todrop_again)

print(loan.shape)
```

```python
print(loan.select_dtypes(include=['object']).columns.tolist())
```

```python
# Label encoding
loan['term'] = loan['term'].str.strip().map({'36 months': 36, '60 months': 60})

grade_map = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7}
loan['grade'] = loan['grade'].map(grade_map)

loan['sub_grade'] = pd.Categorical(loan['sub_grade']).codes

emp_map = {
    '< 1 year': 0, '1 year': 1, '2 years': 2, '3 years': 3,
    '4 years': 4, '5 years': 5, '6 years': 6, '7 years': 7,
    '8 years': 8, '9 years': 9, '10+ years': 10
}
loan['emp_length'] = loan['emp_length'].map(emp_map)

# Drop addr_state — too many unique values
loan = loan.drop(columns=['addr_state'])

# One-hot encoding for remaining categorical columns
loan = pd.get_dummies(loan,
                      columns=['home_ownership', 'verification_status', 'pymnt_plan',
                               'purpose', 'initial_list_status', 'application_type',
                               'disbursement_method'],
                      drop_first=True)

# Confirm no categorical columns remain
print(loan.select_dtypes(include=['object']).columns.tolist())
print(loan.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score
X = loan.drop(columns=['target'])
Y= loan['target']

X_train ,X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, random_state =42,stratify = Y)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

lr_model = LogisticRegression(max_iter=1000 , class_weight='balanced',random_state=42)
lr_model.fit(X_train_scaled, Y_train)
lr_predictions = lr_model.predict(X_test_scaled)

print(classification_report(Y_test, lr_predictions))
print("ROC-AUC:", roc_auc_score(Y_test, lr_predictions))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(class_weight='balanced',random_state=42,max_depth=10)
dt_model.fit(X_train, Y_train)
dt_predictions = dt_model.predict(X_test)

print(classification_report(Y_test, dt_predictions))
print("ROC-AUC:", roc_auc_score(Y_test, dt_predictions))
```

```python
from sklearn.model_selection import train_test_split

# Sample 20% of training data
X_train_sample, _, Y_train_sample, _ = train_test_split(
    X_train, Y_train,
    test_size=0.8,
    random_state=42,
    stratify=Y_train
)

print(X_train_sample.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=20,
    min_samples_split=10,
    min_samples_leaf=5,
    max_features='sqrt',
    class_weight='balanced_subsample',
    random_state=42,
    n_jobs=-1
)
rf_model.fit(X_train, Y_train)
rf_predictions = rf_model.predict(X_test)
rf_probs = rf_model.predict_proba(X_test)[:, 1]

print(classification_report(Y_test, rf_predictions))
print("ROC-AUC:", roc_auc_score(Y_test, rf_probs))
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.1,
    scale_pos_weight=3,  # handles imbalance: ratio of majority/minority class
    random_state=42,
    n_jobs=-1,
    eval_metric='auc'
)
xgb_model.fit(X_train, Y_train)
xgb_predictions = xgb_model.predict(X_test)
xgb_probs = xgb_model.predict_proba(X_test)[:, 1]

print(classification_report(Y_test, xgb_predictions))
print("ROC-AUC:", roc_auc_score(Y_test, xgb_probs))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

# Create a DataFrame pairing feature names with their importance scores
feature_importance = pd.DataFrame({
    'feature': X_train.columns,
    'importance': xgb_model.feature_importances_
})

# Sort by importance descending and keep top 15
feature_importance = feature_importance.sort_values('importance', ascending=False).head(15)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(feature_importance['feature'], feature_importance['importance'])
plt.gca().invert_yaxis()  # highest importance at top
plt.title('Top 15 Features Driving Loan Default Risk')
plt.xlabel('Importance Score')
plt.tight_layout()
plt.show()

# Print the values as well
print(feature_importance)
```

```python
def early_warning_score(model,input_data):
    """
    model: trained XGBoost model
    input_data: dictionary of feature values for one borrower
    returns: risk probability percentage and risk flag
    """
    # Convert input dictionary to DataFrame
    input_df = pd.DataFrame([input_data])
    # Align columns with training data — fills missing columns with 0
    input_df = input_df.reindex(columns=X_train.columns, fill_value=0)
    # Get probability of default (class 1)
    risk_prob = model.predict_proba(input_df)[:,1][0]

    # Flag based on probability threshold
    if risk_prob >= 0.5:
        risk_flag = "HIGH RISK — Flag for review"
    elif risk_prob >= 0.3:
        risk_flag = "MEDIUM RISK — Monitor closely"
    else:
        risk_flag = "LOW RISK — Likely safe"

    return round(risk_prob * 100, 2), risk_flag

# Test with a high risk borrower — Grade E, high DTI, small business loan
high_risk_borrower = {
    'grade': 5,           # Grade E
    'sub_grade': 20,      # E5
    'int_rate': 22.0,     # high interest rate
    'dti': 35.0,          # very high debt-to-income
    'term': 60,           # 60 month loan
    'annual_inc': 40000,  # low income
    'purpose_small_business': 1  # small business loan
}
# Test with a low risk borrower — Grade A, low DTI, credit card loan
low_risk_borrower = {
    'grade': 1,           # Grade A
    'sub_grade': 2,       # A2
    'int_rate': 6.5,      # low interest rate
    'dti': 8.0,           # low debt-to-income
    'term': 36,           # 36 month loan
    'annual_inc': 120000, # high income
    'purpose_small_business': 0
}
# Run both through the function
prob1, flag1 = early_warning_score(xgb_model, high_risk_borrower)
print(f"High Risk Borrower: {prob1}% default probability — {flag1}")

prob2, flag2 = early_warning_score(xgb_model, low_risk_borrower)
print(f"Low Risk Borrower: {prob2}% default probability — {flag2}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Set style
sns.set_style('whitegrid')
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Loan Default Risk Analysis — Key Findings', fontsize=16, fontweight='bold', y=1.02)

# Plot 1: Default rate by grade
grade_data = pd.read_sql_query("""
    SELECT grade, ROUND(AVG(target)*100, 2) as default_rate
    FROM loans
    GROUP BY grade
    ORDER BY grade
""", conn)
axes[0, 0].bar(grade_data['grade'], grade_data['default_rate'],
               color=['#2ecc71', '#27ae60', '#f39c12', '#e67e22', '#e74c3c', '#c0392b', '#922b21'])
axes[0, 0].set_title('Default Rate by Loan Grade', fontweight='bold')
axes[0, 0].set_xlabel('Loan Grade')
axes[0, 0].set_ylabel('Default Rate (%)')

# Plot 2: DTI bucket vs default rate
dti_data = pd.read_sql_query("""
    SELECT
        CASE
            WHEN dti < 10 THEN '1. Low (0-10)'
            WHEN dti BETWEEN 10 AND 20 THEN '2. Medium (10-20)'
            WHEN dti BETWEEN 20 AND 30 THEN '3. High (20-30)'
            ELSE '4. Very High (30+)'
        END as dti_bucket,
        ROUND(AVG(target)*100, 2) as default_rate
    FROM loans
    GROUP BY dti_bucket
    ORDER BY dti_bucket
""", conn)

axes[0, 1].bar(dti_data['dti_bucket'], dti_data['default_rate'],
               color=['#2ecc71', '#f39c12', '#e67e22', '#e74c3c'])
axes[0, 1].set_title('Default Rate by DTI Bucket', fontweight='bold')
axes[0, 1].set_xlabel('DTI Range')
axes[0, 1].set_ylabel('Default Rate (%)')
axes[0, 1].tick_params(axis='x', rotation=15)
for i, v in enumerate(dti_data['default_rate']):
    axes[0, 1].text(i, v + 0.3, f'{v}%', ha='center', fontweight='bold')

# Plot 3: Feature importance (top 10)
feature_importance_plot = feature_importance.head(10)
axes[1, 0].barh(feature_importance_plot['feature'],
                feature_importance_plot['importance'],
                color='#3498db')
axes[1, 0].invert_yaxis()
axes[1, 0].set_title('Top 10 Features Driving Default Risk', fontweight='bold')
axes[1, 0].set_xlabel('Importance Score')

# Plot 4: Interest rate distribution — defaulters vs non-defaulters
defaulters = loan[loan['target'] == 1]['int_rate']
non_defaulters = loan[loan['target'] == 0]['int_rate']

axes[1, 1].hist(non_defaulters, bins=50, alpha=0.6, color='#2ecc71',
                label='Safe (0)', density=True)
axes[1, 1].hist(defaulters, bins=50, alpha=0.6, color='#e74c3c',
                label='Default (1)', density=True)
axes[1, 1].set_title('Interest Rate Distribution: Safe vs Default', fontweight='bold')
axes[1, 1].set_xlabel('Interest Rate (%)')
axes[1, 1].set_ylabel('Density')
axes[1, 1].legend()

plt.tight_layout()
plt.savefig('eda_findings.png', dpi=150, bbox_inches='tight')
plt.show()

print("Visualization saved as eda_findings.png")
```
